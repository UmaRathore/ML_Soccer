```python
import pandas as pd
data_frame = pd.read_csv("archive/players_16.csv")
data_frame
```

```python
df1 = pd.DataFrame(data_frame, columns = ['short_name', 'age', 'value_eur', 'wage_eur'])

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df1['wage_eur'] = df1['wage_eur'].apply(value_to_float)
df1['value_eur'] = df1['value_eur'].apply(value_to_float)

df1['Difference'] = df1['value_eur'] - df1['wage_eur']
df1

df1.sort_values('Difference', ascending=False)
```

```python
import seaborn as sns
sns.set()
graph = sns.scatterplot(x='wage_eur', y='value_eur', data='df1')
graph
```

```python
import pandas as pd
data_frame = pd.read_csv("archive/players_16.csv")
df1 = pd.DataFrame(data_frame, columns = ['short_name', 'age', 'value_eur', 'wage_eur'])

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = df1['wage_eur'].apply(value_to_float)
value = df1['value_eur'].apply(value_to_float)

df1['wage'] = wage
df1['value'] = value

df1['Difference'] = df1['value'] - df1['wage']

df1.sort_values('Difference', ascending=False)
df1
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='wage', y='value', data='df1')
```
